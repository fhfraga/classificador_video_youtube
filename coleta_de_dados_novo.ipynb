```python
import youtube_dlc
import pandas as pd
import numpy as np
```

```python
queries = ["ciencia+de+dados", "machine+learning", "inteligencia+artificial"]
```

```python
ydl = youtube_dlc.YoutubeDL({'ignoreerros': True})
```

```python
resultados = []
for query in queries:
    resultado = ydl.extract_info("ytsearchdate5000:{}".format(query), download=False)
    for entrada in resultado['entries']:
        if entrada is not None:
            entrada['query'] = query
    resultados += resultado['entries']
```

```python
print(len(resultados))
resultados2 = [r for r in resultados if r is not None]
print(len(resultados2))
```

```python
df = pd.DataFrame(resultados)
df.head(2)
```
