# Coleta dos dados

```python
queries = [data science, data engineer, deep learning, inteligencia artificial,
           machine learning, kaggle, series temporais, modelos econometricos, 
           modelos estatisticos, redes neurais recorrentes, 
           transformers deep learning, large language models, ia generativa, 
           redes neurais convulacionais, redes neurais artificiais, arquitetura de dados]
url = 'https://www.youtube.com/results?search_query={query}'
for query in queries:
    query = query.replace(' ', '+')
    urll = url.format(query=query)
    print(query)
    print(urll)
    response = requests.get(urll)
    queries.append(query)
    
    with open(f'./dados_brutos/pesquisa/{query}.html', 'w+') as output:
             output.write(response.text)
    #time.sleep(2)
```

# Processamento dos dados brutos

```python
for query in queries:
    with open(f'./dados_brutos/pesquisa/{query}.html', 'r+') as inp:
        page_html = inp.read()
        
        soup = BeautifulSoup(page_html, 'html.parser')
        parsed = str(soup)
        
        links = re.findall(r'/watch.+?(?=")', parsed)
        
        url = 'https://www.youtube.com{link}'
        for link in links:
            #print(link)
            urll = url.format(link=link)
            print(urll)
            response = requests.get(urll)
            
            try:
                link_name = re.search('v=(.*)', link).group(1)
                with open(f'./dados_brutos/videos/video_{link_name}.html', 'w+') as output:
                    output.write(response.text)
                    #time.sleep(1)
                    print(link_name)
            except:
                link_name = 'Não é um vídeo'
```

```python
with open("dados/parsed_video_info.json", 'w+') as output:
    for video_file in tqdm.tqdm_notebook(sorted(glob.glob("./dados_brutos/videos/video*"))):
        with open(video_file, 'r+') as inp:
            page_html = inp.read()
            soup = BeautifulSoup(page_html, 'html.parser')

            class_watch = soup.find_all(attrs={"class":re.compile(r"watch")})
            id_watch = soup.find_all(attrs={"id":re.compile(r"watch")})
            channel = soup.find_all("a", attrs={"href":re.compile(r"channel")})
            meta = soup.find_all("meta")

            parsed = str(soup)
            
            # visualizacoes
            visualizacoes = re.findall(r'"viewCountText":{"simpleText".+?(?=,)', parsed)
            try:
                numero_visualizacoes = float((visualizacoes[0][31:-16]).replace('.', ''))
            except:
                numero_visualizacoes = None
                
            # inscritos
            inscritos = re.findall(r'"subscriberCountText".+?(?=inscritos")', parsed)
            try:
                if (inscritos[0][74:]).split()[0] == 'mil':
                    numero_inscritos = float((inscritos[0][70:-12]).replace(',', '.')) * 1000 / 1000000
                else:
                    numero_inscritos = float((inscritos[0][70:-12]).replace(',', '.'))
            except:
                numero_inscritos = None
            
            # canal
            canal = re.findall(r'"itemListElement".+?(?="}}]})', parsed)
            canal = str(canal).replace(',', '').replace('"', '').split()
            
            dicionario = {}
            for i, k in enumerate(canal):
                dicionario.update({i:k})
            
            df = pd.DataFrame(dicionario, index=range(len(dicionario))).T
            df_tratado = df[9:]
            lista = []
            for i in df_tratado[0]:
                lista.append(i)
            canal_tratado = ' '.join(lista)
            canal_tratado = canal_tratado.replace("']", '')
            
            # likes
            likes = re.findall(r'"simpleText":"Marcações.+?(?=marcações)', parsed)
            likes = str(likes).replace('"', '').split(':')
            
            dicionario = {}
            for i, k in enumerate(likes):
                dicionario.update({i:k})
            
            df = pd.DataFrame(dicionario, index=range(len(dicionario))).T
            df_tratado = df[2:]
            
            lista = []
            for i in df_tratado[0]:
                lista.append(i)
            likes_tratado = ' '.join(lista)
            likes_tratado = likes_tratado.replace("']", '').replace(',', '.')
            likes_tratado = likes_tratado.split(' ')
            try:
                if likes_tratado[1] == 'mil':
                    likes_tratado = float(likes_tratado[0])
                else:
                    likes_tratado = float(likes_tratado[0]) / 1000
            except:
                likes_tratado = None
                
            # tempo
            date = re.findall(r'"dateText.+?(?="})', parsed)
            date = str(date).replace('"', '').split(':')
            date = (date[2]).replace("']", "").replace('.', '').replace('de', '').strip()
            date = date.split()
             
            data = {}

            for classe in class_watch:
                colname = "_".join(classe['class'])
                if "clearfix" in colname:
                    continue
                data[colname] = classe.text.strip()
                

            for identificador in id_watch:
                colname = identificador['id']
                data[colname] = identificador.text.strip()
                

            for m in meta:
                len(meta)
                colname = m.get('property')
                if colname is not None:
                    data[colname] = m['content']
            
            data['visualizacoes'] = numero_visualizacoes
            data['inscritos_canal_(em_milhoes)'] = numero_inscritos
            data['nome_canal'] = canal_tratado
            data['likes_video_(em_milhares)'] = likes_tratado
            if date[0].lower() == 'estreou':
                data['dia'] = date[2]
                data['mes'] = date[3]
                data['ano'] = date[4]
            elif date[0].lower() == 'transmitido':
                data['dia'] = date[4]
                data['mes'] = date[5]
                data['ano'] = date[6]
            else:
                data['dia'] = date[0]
                data['mes'] = date[1]
                data['ano'] = date[2]

            output.write("{}\n".format(json.dumps(data)))
```

```python
df_tratamento = df[['og:title', 'visualizacoes']]
```

```python
df_table = df_tratamento.pivot_table(values='visualizacoes', aggfunc='mean', index='og:title')
df_table = df_table.reset_index()
```

```python
df_tratamento2 = df_tratamento
df_tratamento2['index'] = df_tratamento2.index
df_tratamento2 = df_tratamento2.drop('visualizacoes', axis=1)
df_table2 = df_tratamento2.pivot_table(values='index', aggfunc='min', index='og:title')
df_table2 = df_table2.reset_index()
```

```python
df_tratada_final = df_table.merge(df_table2, how='left', on='og:title')
```

```python
df['index'] = df.index
df = df.drop('visualizacoes', axis=1)
```

```python
df_final = df.merge(df_tratada_final, how='right', on=['og:title', 'index'])
df_final = df_final.drop('index', axis=1)
df_final = df_final.sample(frac=1)
df_final
```
